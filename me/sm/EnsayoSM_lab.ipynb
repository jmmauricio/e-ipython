```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Ensayo de vacío


```python
R_s=6.5;


I_f=np.arange(0.01,0.45,0.05)
U_s_crece=np.array([8.0, 81, 172, 252, 327, 370, 412, 451, 481])
U_s_decrece=np.array([3, 87, 184, 260, 333, 386, 425, 460, 481])
I_s=np.array([0.05, 0.21, 0.42, 0.62, 0.85, 1.09, 1.25, 1.38, 1.6])
E_f=(U_s_crece+U_s_decrece)/2




fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(I_f, U_s_crece)
ax.plot(I_f, U_s_decrece)

fig.show()

```

### Ensayo de cortocircuito


```python
I_s=np.array([0.05, 0.21, 0.42, 0.62, 0.85, 1.09, 1.25, 1.38, 1.6])
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(I_f, I_s)


fig.show()

```

### Calculo de impedancia síncrona

```python
Z_s=E_f/(np.sqrt(3)*I_s)
X_s= np.sqrt(Z_s**2 - R_s**2)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(I_f, Z_s)
ax.plot(I_f, X_s)
ax.plot(I_f, R_s*np.ones(X_s.shape))
fig.show()
```
