```python
import folium
lat_min, lon_min =  -30.827,  -72.114
lat_max, lon_max =  -16.941 , -66.423

pos = [(lat_min+lat_max)*0.5, (lon_min+lon_max)*0.5]
map_osm = folium.Map(location=pos)
map_osm.create_map(path='osm.html')
stamen = folium.Map(location=pos, tiles='Stamen Toner',
                    zoom_start=8)
stamen.create_map(path='stamen_toner.html')
```

```python
map_1 = folium.Map(location=pos, zoom_start=12,
                   tiles='Stamen Terrain')
map_1.simple_marker([-23.060148,-70.36726], popup='Mt. Hood Meadows')
map_1.create_map(path='mthood.html')

map_1 = folium.Map(location=pos, zoom_start=12,tiles='Stamen Terrain')
map_1.simple_marker([-23.060148,-70.36726], popup='Mt. Hood Meadows',marker_icon='cloud')
map_1.create_map(path='iconTest.html')
```
