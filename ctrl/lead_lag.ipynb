```python
import numpy as np
import scipy.signal as sps
import control as ctrl
```

### Lead-lag compensator

$$\sf G_l = \frac{a Ts + 1}{a Ts + a}$$


$$\sf a = \frac{ 1 + \sin{\left (\phi \right )}}{1 - \sin{\left (\phi \right )} }$$

$$\sf T = \frac{ 1 }{\omega \sqrt{a} }$$

```python
T_ag = 0.5
G_ag = ctrl.tf(1,[T_ag, 1])

omega = 2.0*np.pi*1.0
phase = np.angle(G_ag.evalfr(omega)[0][0])


phi = -phase
a = (1.0+np.sin(phi))/(1.0-np.sin(phi))
T = 1.0/(omega*np.sqrt(a))



G_lead = ctrl.tf([a*T, 1],[a*T, a])

G_ag_lead = ctrl.series(G_ag,G_lead)


mag_ag,      phase_ag,      omega_ag      = ctrl.bode(G_ag, Plot=False)
mag_lead,    phase_lead,    omega_lead    = ctrl.bode(G_lead, Plot=False)
mag_ag_lead, phase_ag_lead, omega_ag_lead = ctrl.bode(G_ag_lead, Plot=False)
```

```python
%matplotlib inline   
import matplotlib.pyplot as plt

fig_bodes = plt.figure(figsize=(10,8))

ax_m     = fig_bodes.add_subplot(2,1,1)
ax_phase = fig_bodes.add_subplot(2,1,2)

ax_m.semilogx(omega_ag,  mag_ag, label='AVR+GEN')    # bode phase plot
ax_m.semilogx(omega_lead,mag_lead, label='COMP.')  # bode phase plot
ax_m.semilogx(omega_ag_lead,mag_ag_lead, label='COMP.+AVR+GEN')  # bode phase plot

ax_phase.semilogx(omega_ag, phase_ag, label='AVR+GEN')    # bode phase plot
ax_phase.semilogx(omega_lead, phase_lead, label='COMP.')  # bode phase plot
ax_phase.semilogx(omega_ag_lead, phase_ag_lead, label='COMP.+AVR+GEN')  # bode phase plot

ax_m.legend()
ax_phase.legend()


ax_phase.semilogx([omega, omega],[-100.0, 100.0])    # bode phase plot
ax_m.grid(True)
ax_phase.grid(True)

plt.show()
```

```python
T_w = 2.0
G_wash_out = G_lead = ctrl.tf([T_w, 0],[T_w, 1])

K_stab = a

G_stab = ctrl.tf([K_stab],[1])

G_pss = ctrl.series(G_stab, ctrl.series(G_wash_out, G_lead))
G_total = ctrl.series(G_pss, G_ag)
```

```python
mag_pss,      phase_pss,      omega_pss      = ctrl.bode(G_pss, Plot=False)
mag_total,    phase_total,    omega_total    = ctrl.bode(G_total, Plot=False)

%matplotlib inline 
fig_bodes_final = plt.figure(figsize=(10,8))

ax_m     = fig_bodes_final.add_subplot(2,1,1)
ax_phase = fig_bodes_final.add_subplot(2,1,2)

ax_m.semilogx(omega_pss,  mag_pss, label='PSS')    # bode phase plot
ax_m.semilogx(omega_total,mag_total, label='Final')  # bode phase plot

ax_phase.semilogx(omega_pss,  phase_pss, label='PSS')    # bode phase plot
ax_phase.semilogx(omega_total,phase_total, label='Final')  # bode phase plot

ax_m.legend()
ax_phase.legend()

ax_m.semilogx([omega, omega],[-10.0, 50.0])    # bode phase plot
ax_phase.semilogx([omega, omega],[-100.0, 100.0])    # bode phase plot
ax_m.grid(True)
ax_phase.grid(True)

plt.show()
```
