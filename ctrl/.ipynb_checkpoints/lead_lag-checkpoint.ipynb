```python
import numpy as np
import scipy.signal as sps
```

```python
T = 1.0
a = 2.0
w_n = 1.0/(T*np.sqrt(a))
numlead=np.array([a*T, 1])
denlead=np.array([a*T, a])
G_lead = sps.lti(numlead,denlead)
phi  =np.arcsin((a-1)/(a+1))
180.0/np.pi*phi, w_n
```

```python
%matplotlib inline   
import matplotlib.pyplot as plt
w, mag, phase = sps.bode(G_lead)

plt.figure()
plt.semilogx(w, mag)    # bode magnitude plot
plt.grid(True)

plt.figure()
plt.semilogx(w, phase)  # bode phase plot
plt.semilogx(w_n, 180.0/np.pi*phi, 'ro')  # bode phase plot
plt.grid(True)
plt.show()

```
