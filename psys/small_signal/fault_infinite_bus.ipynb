```python
from pysimu import sim
import numpy as np
 
simu_1 = sim()
```

## Generator to infinite bus

### Data

```python
p_m = 1.0
X = 0.5
e = 1.0
v = 1.0
H = 3.5
omega_s = 1.0
omega = omega_s
D = 1.0
Omega_b = 2.0*np.pi*50.0
```

```python
def f(t,x):
    
    delta = x[0]
    omega = x[1]
    
    p_e = e*v/X*np.sin(delta)
    
    ddelta = Omega_b*(omega - omega_s)
    domega = 1.0/(2*H)*(p_m - p_e - D*(omega - omega_s))
    
    return [ddelta, domega]


def h(t,x):

    delta = x[0]
    omega = x[1]
    
    p_e = e*v/X*np.sin(delta)
    
    return np.array(p_e)
```

### Initial values

```python
p_e = p_m
delta_0 = np.arcsin(p_e*X/(e*v))  
omega_0 = omega_s
```

### Simulation

```python
x_0 = np.array([delta_0, omega_0])

simu_1.f = f
simu_1.x_0 = x_0
simu_1.h = h

simu_1.run(1.0)

v = 0.0
fault_ms = 50.0

simu_1.run(1.0+fault_ms/1000.0)

v = 1.0

simu_1.x_0 = simu_1.x
simu_1.run(5.0)
```

### Plots
